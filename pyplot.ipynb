### Plotting with keyword strings


There are some instances where you have data in a format that lets you access particular variables with strings. For example, with numpy.recarray or pandas.DataFrame.

Matplotlib allows you provide such an object with the data keyword argument. If provided, then you may generate plots with the strings corresponding to these variables.



```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

### Plotting with categorical variables

It is also possible to create a plot using categorical variables. Matplotlib allows you to pass categorical variables directly to many plotting functions.



```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

### Annotating text
The uses of the basic text function above place text at an arbitrary position on the Axes. A common use for text is to annotate some feature of the plot, and the annotate method provides helper functionality to make annotations easy. In an annotation, there are two points to consider: the location being annotated represented by the argument xy and the location of the text xytext. Both of these arguments are (x, y) tuples.

```python
ax = plt.subplot()

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             )

plt.ylim(-2, 2)
plt.show()
```

```python
# lets use shorthand notation for the graph
# fmt ='[color][marker][line]'

x= [1,2,3]
y= [2,4,6]

#plt.plot(x,y, label='2x', color = 'red', linewidth=2, marker='.', linestyle='--',markersize=12, markeredgecolor='green')
plt.plot(x,y, 'r^--', label = '2x')
plt.title('Basic Graph on Matplotlib', fontdict={'fontname': 'serif', 'fontsize': 15})
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```python
# now we will do a second line in the graph using the numpy function 

x= [0,1,2,3,4]
y= [0,2,4,6,8]
#Resize the graph and dpi. Dpi specifies pixels per inch. 
plt.figure(figsize=(7,4), dpi=100)
#plt.plot(x,y, label='2x', color = 'red', linewidth=2, marker='.', linestyle='--',markersize=12, markeredgecolor='green')
plt.plot(x,y, 'r^--',label='2x')

#select the interval we want to plot the points at 
x2 = np.arange(0,4.5,0.5)
#plot part of the graph as a line and then the remainder as --
plt.plot(x2[:6], x2[:6]**2, 'g', label='X^2') # the 6 here is exlusive that is why we started at five for the line below
plt.plot(x2[5:], x2[5:]**2, 'g--')

plt.title('Basic Graph on Matplotlib', fontdict={'fontname': 'serif', 'fontsize': 15})
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xticks([0,1,2,3,4,])
plt.legend()
plt.savefig('practicegraph01.png', dpi=100)
plt.show()
```

### Working with real world data
Data downloaded from a github tutorial account
ref: https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/master/gas_prices.csv

```python
gas = pd.read_csv('data/gas_prices.csv')
gas
```

```python
plt.plot(gas.Year, gas.USA) # x is year and y axis is gas price
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Gas prices in US Dollars')

plt.plot(gas.Year, gas.USA, 'b.-',label= 'US') # x is year and y axis is gas price
plt.plot(gas.Year, gas.Canada, 'r^-' ,label= 'Canada')
plt.plot(gas.Year, gas['South Korea'],'g*-' ,label = 'SK') # need to use a list here because of the space in the name in South Korea
plt.plot(gas.Year, gas.France, 'y^-', label ='France')
plt.xlabel('Year')
plt.ylabel('US Dollar')
plt.xticks(gas.Year[::2]) # only want every two years as the figure will be too squashed if we do every year
plt.legend()       
plt.show()
```

### Lets Visualize all the countries on the plot:


```python
plt.figure(figsize=(10,5))
plt.title('Gas prices in US Dollars', fontdict={'fontweight': 'bold', 'fontsize': 15})

for country in gas:
    if country!= 'Year':
        plt.plot(gas.Year, gas[country], marker = '.', label=country)
plt.xlabel('Year')
plt.ylabel('US Dollar')
plt.xticks(gas.Year[::2]) # only want every two years as the figure will be too squashed if we do every year
plt.legend()       
plt.show()
```

### Lets look at specific countries using a for loop:


```python
plt.figure(figsize=(10,5))
plt.title('Gas prices in US Dollars',fontdict={'fontweight': 'bold', 'fontsize': 15})
countries_to_look_at =['France','Germany', 'Italy', 'UK']
for country in gas:
    if country in countries_to_look_at:
        plt.plot(gas.Year, gas[country], marker = '.', label=country)
plt.xlabel('Year')
plt.ylabel('US Dollar')
plt.xticks(gas.Year[::2]) # only want every two years as the figure will be too squashed if we do every year
plt.legend()       
plt.show()
```

The plot above shows up that gas prices have been consistently going up within all the EU countries in this time period.
