```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Playing with matploylib.pyplot and looking at the functions [1]

```python
fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.plot([1, 2, 3, 4], [1, 4, 8, 13])  # Plot some data on the axes.
```

Looking at Matplotlib Markers [2]

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'D')
plt.show()
```

Reviewing matplotlib videos in research on plots. Python: Intro to Visualization with Matplotlib[3]


```python
ts = pd.Series(np.random.randn(180), index=pd.date_range('1/1/2021', periods=180))
df = pd.DataFrame(np.random.randn(180, 3), index=ts.index, columns=list('ABC'))
df.cumsum().plot()
plt.show()
```
