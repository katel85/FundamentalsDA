```python
def points_to_array(s):
    portfolio = ''
    if s[0] =='#':
        portfolio = '#'
    random = ''
    if s[-1] =='*':
        random = '*'
    points = ''
    for i in s:
        if i.isdigit():
            points = points + i
    return [points,portfolio,random]
```
